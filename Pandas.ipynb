```python
########### DATA CLEANING AND PREPROCESSING *****************
## *************** vital step before extracting insights
```

```python
import pandas as pd

df = pd.read_csv("/content/Customers.csv")
df.head()
```

```python
# Removing rows with empty cells - DROPNA()

new_df = df.dropna()
print(new_df.to_string())
```

```python
df.fillna(100, inplace= True) # REPLACING THE NULL COLUMNS WITH A VALUE , ***ISSUE - Whereever there's null value, replaces with 100 whether it be a string record too
print(df.to_string())


```

```python
df.fillna({"Company": "NoName"}, inplace=True) # if we wanna replace a specific column with some value, having null
```

```python
# REMOVING DUPLICATES

df_new = df.drop_duplicates()
```

```python
df["Subscription Date"] = pd.to_datetime(df['Subscription Date'], format = "%Y-%m-%d")
df.head()
```

```python
print(df['Subscription Date'].dtype)
```

```python
### *************** DATA MANIPULATION AND AGGREGATION *******************

customers_per_city = df.groupby('Country')['Customer Id'].value_counts()
print(customers_per_city)
```

```python
## ________________ DATA VISUALIZATION __________

import pandas as pd
import matplotlib.pyplot as plt


data = {'Date': ['2023-08-01', '2023-08-02', '2023-08-03'], 'Price': [100, 105, 110]}

df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.plot(x = 'Date', y = 'Price', kind = 'line')
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

```python
data = {
    'Product': ['A', 'B', 'C'],
    'Sales': [500, 750, 300]
}
df = pd.DataFrame(data)
```

```python
# Barchart

df.plot(x = "Product", y = "Sales", kind = "bar")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```

```python
data = {
    'Hours_Studied': [2,4,5,7,8,10],
    'Exam_Score': [60, 70, 75, 85, 90, 95]
}

df = pd.DataFrame(data)
```

```python
# Scatter Plot

df.plot(x = "Hours_Studied", y = "Exam_Score", kind = 'scatter')
plt.show()
```

```python
# Pie chart

data = {
    'Topping': ['Pepperoni', 'Mushroom', 'Olive', 'Sausage'],
    'Count': [25, 15, 10, 20]
}
df = pd.DataFrame(data)
```

```python
df.plot(x = "Topping", y = "Count", kind= 'pie')
plt.show()
```
